```python
def cabeceras (dataframe):

    """Esta función renombra las cabeceras del DataFrame, sustituye espacios por "_"
    y las pone en minusculas.
    
    Args:
    lista : lista de DFs que queremos modificar
    
    Returns:
    La funcion no tiene return, modifica las cabeceras por asignacion directa"""

    dataframe.columns = [col.strip().replace(" ", "_").lower() for col in dataframe.columns]
    print(f"Las nuevas cabeceras son: {dataframe.columns}")
    print("--------------------------------------------")
```

```python
afflicted = []
age = []
residence = []
info = []

# Procesamiento de cada entrada
for entry in data:
    # Separar la información
    name_match = re.match(r"(.+?)\s*[,–-]\s*age", entry)
    name = name_match.group(1).strip() if name_match else None
    
    age_match = re.search(r"age(?: about)? (\d+)", entry)
    age_value = int(age_match.group(1)) if age_match else None
    
    residence_match = re.search(r"living in ([^.]+)", entry)
    residence_value = residence_match.group(1).strip() if residence_match else None
    
    info_match = re.search(r"living in (.+)\. (.+)", entry)
    info_value = info_match.group(2).strip() if info_match else None
    
    afflicted.append(name)
    age.append(age_value)
    residence.append(residence_value)
    info.append(info_value)

# Crear el DataFrame
acusados = pd.DataFrame({
    'afflicted': afflicted,
    'age': age,
    'residence': residence,
    'info': info
})
```

```python
def process_entry(entry):
    # Separar nombre, edad y residencia
    name_match = re.match(r"(.+?)\s*[,–-]\s*age", entry)
    name = name_match.group(1).strip() if name_match else None
    
    age_match = re.search(r"age(?: about)? (\d+)", entry)
    age = int(age_match.group(1)) if age_match else None
    
    # Extraer residencia
    residence_match = re.search(r"living in ([^.]+)", entry)
    residence = residence_match.group(1).strip() if residence_match else None
    
    return {'afflicted': name, 'residence': residence, 'age': age}
```
