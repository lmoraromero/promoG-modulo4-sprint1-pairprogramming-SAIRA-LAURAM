```python
def cabeceras (dataframe):

    """Esta funci√≥n renombra las cabeceras del DataFrame, sustituye espacios por "_"
    y las pone en minusculas.
    
    Args:
    lista : lista de DFs que queremos modificar
    
    Returns:
    La funcion no tiene return, modifica las cabeceras por asignacion directa"""

    dataframe.columns = [col.replace(".", "_").lower() for col in dataframe.columns]
    print(f"Las nuevas cabeceras son: {dataframe.columns}")
    print("--------------------------------------------")


cabeceras(df_general_trials)
```

```python
df_general_trials["year"] = df_general_trials["year"].apply(lambda dato: str(dato).split(".")[0])
```

```python
df_general_trials["year"] = df_general_trials["year"].replace("nan", "Desconocido")
```

```python
lista = ["city", "gadm_adm2", "gadm_adm1"]
for col in lista:
    df_general_trials[col] = df_general_trials[col].replace(np.nan, "Desconocido")
```
